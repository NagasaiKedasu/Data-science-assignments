## **Preprocess the Data:**

```python
#Check for null values
data.isnull().sum()
```

```python
# Drop the 'animal name' column as it is not needed for analysis
data = data.drop(columns=['animal name'])
```

## **Split the dataset into training and testing sets (80% training, 20% testing).**

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X = data.drop(columns=['type'])
y = data['type']

# Split the dataset into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

## **Implement the K-Nearest Neighbors algorithm:**

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model on the training data
knn.fit(X_train, y_train)
```

## **Choose an Appropriate Distance Metric and K Value:**

```python
# Use Euclidean distance
knn_euclidean = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn_euclidean.fit(X_train, y_train)

# Use Manhattan distance
knn_manhattan = KNeighborsClassifier(n_neighbors=3, metric='manhattan')
knn_manhattan.fit(X_train, y_train)
```

## **Evaluate the Classifier's Performance:**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predict the labels on the test set
y_pred = knn.predict(X_test)

# Evaluate the classifier's performance
accuracy = accuracy_score(y_test, y_pred)*100
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```

## **Visualize the Decision Boundaries:**

```python
from matplotlib.colors import ListedColormap
import numpy as np

# Use only two features for visualization
X_train_vis = X_train[['hair', 'feathers']]
X_test_vis = X_test[['hair', 'feathers']]
knn_vis = KNeighborsClassifier(n_neighbors=3)
knn_vis.fit(X_train_vis, y_train)

# Create a mesh grid for plotting decision boundaries
x_min, x_max = X_train_vis.iloc[:, 0].min() - 1, X_train_vis.iloc[:, 0].max() + 1
y_min, y_max = X_train_vis.iloc[:, 1].min() - 1, X_train_vis.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict over the grid
Z = knn_vis.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundaries and scatter plot
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(('red', 'skyblue', 'blue', 'yellow', 'purple', 'orange', 'pink')))
plt.scatter(X_train_vis.iloc[:, 0], X_train_vis.iloc[:, 1], c=y_train, s=20, edgecolor='k')
plt.title('Decision Boundaries of KNN Classifier')
plt.xlabel('Hair')
plt.ylabel('Feathers')
plt.show()
```

## **Interview Questions:**

**1. What are the key hyperparameters in KNN?**

The key hyperparameters in the K-Nearest Neighbors (KNN) algorithm are:

**Number of Neighbors (K):**

This is the most important hyperparameter in KNN. It defines the number of nearest neighbors to consider when making a prediction.
A small K (e.g., 1 or 3) makes the model more sensitive to noise, potentially leading to overfitting.
A large K value makes the model more generalized, considering more neighbors, but may lead to underfitting.

**Distance Metric:**

This hyperparameter determines how the distance between data points is calculated.                                    Common distance metrics included are     
  **Euclidean Distance:** The most common metric, which measures the straight-line distance between two points.  
**Manhattan Distance:** Measures the distance between two points along axes at right angles (also known as L1 norm).    
**Minkowski Distance:** A generalization that includes both Euclidean and Manhattan distances as special cases, controlled by a parameter p.


**2. What distance metrics can be used in KNN?**

In K-Nearest Neighbors (KNN), several distance metrics can be used to measure the similarity between instances. The choice of distance metric can significantly affect the performance of the KNN algorithm. Here are the most commonly used distance metrics:

1. **Euclidean Distance:**
Euclidean distance is the most common distance metric used in KNN. It measures the straight-line (or "as-the-crow-flies") distance between two points in Euclidean space.  
Use Case: Best suited for continuous, real-valued data where the relationships between features are linear.
2. **Manhattan Distance (L1 Norm):**
Also known as "Taxicab" or "City Block" distance, Manhattan distance measures the distance between two points by summing the absolute differences of their coordinates.   
Use Case: Useful when the features are not correlated or when you want to minimize the effect of outliers.
3. **Minkowski Distance:**
Minkowski distance is a generalized metric that includes both Euclidean and Manhattan distances as special cases. The parameter p defines the norm of the distance.
p = 1: Manhattan distance
p = 2: Euclidean distance   
Use Case: When you want flexibility in choosing between Manhattan and Euclidean distances, or exploring other intermediate norms
