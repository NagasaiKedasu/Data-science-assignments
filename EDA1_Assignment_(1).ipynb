## **Data Cleaning:**



### **1.Check for Null values**





```python
df.isnull()
```

```python
#Lets sumup the null values
df.isnull().sum()
```

from the above we can notice that there are null values in columns named "LB","AC","DS","DP","MLTV","Width","Tendency","NSP".



### **2.Handle those Null values**



We can handle those null or missing values by using various approches, One of them was Data Imputation.Lets fill the null value with the mean value respectively.

```python
df['LB']=df['LB'].fillna(df['LB'].mean())
df['AC']=df['AC'].fillna(df['AC'].mean())
df['DS']=df['DS'].fillna(df['DS'].mean())
df['DP']=df['DP'].fillna(df['DP'].mean())
df['MLTV']=df['MLTV'].fillna(df['MLTV'].mean())
df['Width']=df['Width'].fillna(df['Width'].mean())
df['Tendency']=df['Tendency'].fillna(df['Tendency'].mean())
df['NSP']=df['NSP'].fillna(df['NSP'].mean())
```

```python
#Once again check for null values
df.isnull().sum()
```

### **3.Check for Duplicate values and drop them**

```python
df.duplicated().sum()
```

```python
#drop the duplicate values
df2=df.drop_duplicates()
```

### **4.Check for Outliers**

```python
# Define a function outlier
def outlier(data):
  # Calculate the first quartile (25th percentile)
  q1 = data.quantile(0.25)
  # Calculate the third quartile (75th percentile)
  q3 = data.quantile(0.75)
  # Calculate the interquartile range (IQR)
  iqr = q3 - q1
  # Calculate the lower and upper bounds
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  # Return the lower and upper bounds
  return lower_bound, upper_bound
```

```python
# Initialize dictionaries to store lower and upper bounds for each column
lower_bounds = {}
upper_bounds = {}

# Iterate over each column in the DataFrame
for column in df.columns:
    lower, upper = outlier(df[column])
    lower_bounds[column] = lower
    upper_bounds[column] = upper
```

### **5.Remove Outliers**

```python
# Store the copy of df2 in new_df
new_df = df2.copy()
for column in df2.columns:
    new_df = new_df[(new_df[column] >= lower_bounds[column]) & (new_df[column] <= upper_bounds[column])]
```

```python
new_df
```

```python
new_df.describe()
```

### **6.Drop the Unwanted Columns**

```python
single_value_columns = [col for col in new_df.columns if new_df[col].nunique() == 1]
final_df = new_df.drop(columns=single_value_columns)
# This is our Cleaned DataFrame and ready to visualize
final_df
```
