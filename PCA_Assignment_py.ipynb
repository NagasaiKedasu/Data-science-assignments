## **Dimensionality Reduction with PCA:**

```python
#Standarization
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
#Implement PCA
pca = PCA()
pca.fit(data_scaled)
```

**Scree Plot:** Plot the explained variance for each principal component using a scree plot to determine the optimal number of components

```python
plt.figure(figsize=(15, 10))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

**Cumulative Explained Variance:** Calculate the cumulative explained variance to see how many components are needed to explain a high percentage of the variance.

```python
cumulative_variance = pca.explained_variance_ratio_.cumsum()
plt.figure(figsize=(15, 10))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.show()
```

```python
n_components = 2
pca = PCA(n_components=n_components)
data_pca = pca.fit_transform(data_scaled)
```

## **Clustering with Original Data:**

```python
kmeans = KMeans(n_clusters=4, random_state=42)
clusters_original = kmeans.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=clusters_original, cmap='viridis')
plt.title('K-means Clustering (Original Data)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Evaluate Silhouette Score and Davies Bouldin Score
from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette_original = silhouette_score(data_scaled, clusters_original)
davies_bouldin_original = davies_bouldin_score(data_scaled, clusters_original)

print(f'Silhouette Score (Original Data): {silhouette_original}')
print(f'DaviesBouldin Index (Original Data): {davies_bouldin_original}')
```

##  **Clustering with PCA Data:**

```python
clusters_pca = kmeans.fit_predict(data_pca)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=clusters_pca, cmap='viridis')
plt.title('K-means Clustering (PCA Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
#Evaluate Silhouette Score and Davies Bouldin Score
silhouette_pca = silhouette_score(data_pca, clusters_pca)
davies_bouldin_pca = davies_bouldin_score(data_pca, clusters_pca)

print(f'Silhouette Score (PCA Data): {silhouette_pca}')
print(f'DaviesBouldin Index (PCA Data): {davies_bouldin_pca}')
```

## **Comparison and Analysis:**

```python
print(f'Silhouette Score Comparison: Original Data = {silhouette_original}, PCA Data = {silhouette_pca}')
print(f'DaviesBouldin Index Comparison: Original Data = {davies_bouldin_original}, PCA Data = {davies_bouldin_pca}')
```

The PCA-transformed data has a significantly higher Silhouette Score (0.507) compared to the original data (0.271). This suggests that the clusters formed after PCA are more compact and better separated than those formed using the original dataset   

The PCA-transformed data has a much lower DaviesBouldin Index (0.699) compared to the original data (1.783). This indicates that the clusters in the PCA-transformed dataset are more distinct and have less overlap compared to the original data clusters.

## **Conclusion:**

The results suggest that applying PCA before clustering significantly improved the clustering quality. This demonstrates that dimensionality reduction can be a powerful preprocessing step, especially when dealing with high-dimensional datasets or when the original features are highly correlated.
